```python
# load mask & ROI

lh_file = os.path.join(nsd_dir, "nsddata", "freesurfer", "fsaverage", "label", 'lh.HCP_MMP1.mgz')
rh_file = os.path.join(nsd_dir, "nsddata", "freesurfer", "fsaverage", "label", 'rh.HCP_MMP1.mgz')
maskdata_lh = nib.load(lh_file).get_fdata().squeeze()
maskdata_rh = nib.load(rh_file).get_fdata().squeeze()
maskdata = np.hstack((maskdata_lh, maskdata_rh))

ROI_path = os.path.join(proj_dir, "utils", "ROI_labels.tsv")
ROI_df = pd.read_csv(ROI_path, sep='\t')

```

```python
# make vox x cate matrix

# matmal to get sum of product
vox_cate_mat = np.dot(masked_betas, label_matrix)
# normalize by cate frequency
freq = label_matrix.sum(axis=0)
vox_cate_mat = np.divide(vox_cate_mat, freq, out=np.zeros_like(vox_cate_mat), where=freq!=0)
vox_cate_mat.shape  # vex x 80
```

```python
# map voxels to color space
vox_color_mat = np.dot(vox_cate_mat, pcs[:3].T)
print("vox_color_mat shape:", vox_color_mat.shape)

# normalize to [0,1]
colormap = np.zeros(vox_color_mat.T.shape)
for i in range(3):
    colormap[i] = (vox_color_mat[:,i] - vox_color_mat[:,i].min()) / (vox_color_mat[:,i].max() - vox_color_mat[:,i].min())
print("colormap shape:", colormap.shape)
```

    * load labels for sample trials
    * load betas for each hemi
        * select with ROIs, get ROI_mask_\*h
        * filter for good voxels, get good_vox_mask_\*h
    * concate hemis for pca
        * dot(betas, trial_label_matrix) 
        * normalize by category appearing frequency --> vox_cate_mat
        * PCA on vox_cate_mat --> cate_pcs
        * dot(vox_cate_mat, cate_pcs) --> vox_pcs
        * normalize vox_pcs to \[0, 1\] as RGB colors
* plotting
    * load fsaverage mesh from nilearn
    * plotting.plot_surf_contours
        * parcellation: len = total vox for hemi. label each voxel for parcellation ROIs 
            * our case: diff value for each vox to show diff colors
        * labels: names of the levels selected, for region legend
            * our case: omit 
        * levels=regions_indices: regions within parcellation labels to be plotted
            * out case: combination of ROI_mask & good_vox_mask as voxel index
        * colors: RGB colors for each voxel
            * normalized vox_pcs
        
        

```python
# === make color map

# parcellation: label of each voxel for parcellation ROIs. diff value for diff colors
nonzeros = np.nonzero(vox_color_lh[:,0])[0]
nonzeros_index = np.arange(len(nonzeros))
parcellation = np.zeros(vox_color_lh[:,0].shape)
parcellation[nonzeros] = nonzeros_index

# # labels: names of the labels selected
# labels = ROI

# regions_indices: regions within parcellation labels to be plotted
regions_indices = nonzeros_index

# color for each cate label 
colors = vox_color_lh[nonzeros,:]
```

```python

# all labels
labels = np.load(f"./rsa/all_stims_category_labels_session-10.npy", allow_pickle=True)
all_labels = np.array(sorted(list(set(np.concatenate(labels)))))

# load category label
label_file = os.path.join(betas_dir, f'{sub}_sample_labels_session-{n_sessions}.npy')

if os.path.exists(label_file):
    category_matrix = np.load(label_file, allow_pickle=True)

else:
    # load conditions data
    conditions = get_conditions(nsd_dir, sub, n_sessions)
    conditions = np.asarray(conditions).ravel()  # ntrials x 1
    # then we find the valid trials for which we do have 3 repetitions.
    conditions_bool = [
        True if np.sum(conditions == x) == 3 else False for x in conditions]
    # find the subject's condition list (sample pool)
    sample = np.unique(conditions[conditions_bool])

    category_matrix = get_labels(sub, betas_dir, nsd_dir, sample-1, n_sessions=n_sessions, n_jobs=1)
```

```python
# ===== category cluster pcs ======
class_labels = []
for cate_bool in category_matrix:
    im_lab = all_labels[cate_bool.astype(bool)]
    
    print(im_lab)
    break
```

```python
# ===== NSD =====
# prepare the class labels
class_labels = []
for categ_v in category_matrix:

    # category_dict: 1 is animate, 0 inanimate
    cat_is = np.where(categ_v)[0]
    anim_class = [category_dict[str(x)] for x in cat_is]

    n_anim = np.sum(anim_class)

    # special case only people
    if len(cat_is) == 1 and cat_is == 49:
        class_label = 'a_people'

    # people with other animates
    elif 49 in cat_is and n_anim == len(anim_class):
        class_label = 'a_people_animates'

    # people with inanimates
    elif 49 in cat_is and n_anim == 1:
        class_label = 'a_people_inanimates'

    # people with both animates and inanimates
    elif 49 in cat_is and n_anim < len(anim_class):
        class_label = 'a_people_animates_inanimates'

    # all ones? only animate
    elif n_anim == len(anim_class):
        class_label = 'animates'

    elif np.sum(anim_class) == 0:  # only inanimate
        class_label = 'inanimates'

    # mixed non-people and inanimates
    else:
        class_label = 'animates_inanimates'

    class_labels.append(class_label)
```

```python
roi = "pVTC"
tsne_file = os.path.join(
    outpath, sub, f'{sub}_{roi}_tsne_session-{n_sessions}.npy'
)

if not os.path.exists(tsne_file):
    rdm_file = os.path.join(
        outpath, f'{sub}_{roi}_fullrdm_correlation_session-{n_sessions}.npy'
    )
    print(f'loading full rdm for {roi} : {sub}')
    rdm = np.load(rdm_file, allow_pickle=True)

    # MDS
    print("computng mds...")
    Y_mds = mds(rdm, n_components=2)

    # tsne
    print("computing tsne...")
    tsne_operator = manifold.TSNE(
        metric='precomputed',
        perplexity=100,
        n_components=2,
        init=Y_mds,
        n_jobs=n_jobs
    )
    Y_tsne = tsne_operator.fit_transform(squareform(rdm))

else: 
    print("loading from saved tsne file...")
    Y_tsne = np.load(tsne_file)
```
